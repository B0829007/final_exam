$n$ 個獨立且分布相同的隨機變數 
$
X_1,\ X_2, …. X_n
$
它們有共同的

期望值: $E(X_i)=μ$, 

標準差: $Std(X_i)=σ$, 

----
$\bar{X}$ , $S^2$ 

分別代表它們的【樣本平均數】以及【樣本變異數】。數學公式定義如下：

$
\bar{X}=\frac{1}{n}\sum_{i=1}^{n}X_i
\\
S^2=\frac{1}{n-1}\sum_{i=1}^{n}\left(X_i-\bar{X}\right)^2
$

----
令 統計量 $T$ 代表 樣本平均$\bar{X}$ 與母體平均$μ$ 之間的【正規距離】(scaled distance)，數學公式定義如下：

$
T=\frac{\ \bar{X}-\mu}{S/\sqrt{n}}
$

```python
import numpy      as np
import scipy.special  as sp
import scipy.stats   as st
import pandas      as pd


n= 10 
μ= 5
σ= 2
  
```

1.

	E\left(\bar{X}\right)=\ ?
$


```python
#(1)
```

$E\left(\bar{X}\right)=E(\frac{1}{n}\sum_{i=1}^{n}X_i)=\frac{1}{n}E(\sum_{i=1}^{n}X_i)=\frac{1}{n}\sum_{i=1}^{n}μ=μ=5$


2. 
$
	E\left(S^2\right)=\ ?
$


```python
#(2)
```

$E\left(S^2\right)=E\left[\frac{1}{n-1}\sum_{i=1}^{n}\left(X_i-\bar{X}\right)^2\right]=\frac{1}{n-1}\left[\sum_{i=1}^{n}E(X_i-μ)^2-nE\left(\bar{X}-μ\right)^2\right]=\frac{1}{n-1}\left(nσ^2-nσ_\bar{X}^2\right)=\frac{1}{n-1}\left(nσ^2-n\frac{σ^2}{n}\right)=σ^2=4$

10.

```python
#(10)
import math

μ= 5
σ= 2
n=10
prob1=st.norm.cdf(x=(6-μ)/(σ/math.sqrt(n)))
prob2=st.norm.cdf(x=(4-μ)/(σ/math.sqrt(n)))
prob1-prob2
```

#[Q2]

某大學於某年做了學生身高資料普查，學生總數N為 10000人，全體學生身高平均數μ= 159.8cm，標準差σ = 9.9 cm，資料分布約略如下圖：(橫坐標為身高cm、縱坐標為學生人數)


```python
# 針對某大學全體學生做了一次身高普查，學生總數 N= 10000
# 普查資料 存放在 X 

# 先以模擬方式製造一堆 (size= 10000) 符合此種分布的實驗數據

np.random.seed(0)
X= st.norm.rvs(size=10000, loc= 160, scale=10)

# 畫出 分布圖，
# 並觀察母體平均數 mu= E(X)=μ== 159.8, 
# 以及 標準差 sigma= Std(X)=σ== 9.9

X= pd.Series(X)
X.hist(bins= 100, grid=True)
X.describe().round(1)
```

某同學自己在好朋友之間 (10人) 做了一個調查，得到結果如下：



```python
# 某位本課程的學生作了一個小實驗，
# 他隨機調查了 n=10 位同學 的身高資料
np.random.seed(0)
smpSize= 10
smpX= X.sample(n= smpSize)
smpX.round(1)
aDF= pd.DataFrame(smpX,columns=['身高(cm)'])
aDF.index.name='學號'
aDF.T.round(1)
```

19.	以20人資料重新估算μ的95%信心區間 $(μ_{201}, μ_{202})$=? 

20.	以20人的資料估算出來的信心區間 $(μ_{201}, μ_{202})$ 比10人的資料所估算的信心區間 $(μ_1, μ_2)$ 何者寬度較小？(請寫出各自的信心區間寬度) 


```python
#(19)
import statistics
import math
list1 = [147.6, 185.9, 154.0, 147.2, 168.6, 146.8, 162.6, 164.1, 152.2, 171.8, 151.4, 152.0, 177.1, 155.1, 159.8, 188.0, 163.7, 178.6, 144.0, 161.8]
mean = statistics.mean(list1)
arr=[147.6, 185.9, 154.0, 147.2, 168.6, 146.8, 162.6, 164.1, 152.2, 171.8, 151.4,  152.0, 177.1, 155.1, 159.8, 188.0, 163.7, 178.6, 144.0, 161.8]
arr = statistics.stdev(arr)
n1 = 20
μ202=mean+1.96*(arr/math.sqrt(n1))
μ201=mean-1.96*(arr/math.sqrt(n1))
print(μ201,μ202)
```

```python
#(20)
μ1μ2 = 168.074894 - 152.091866
μ201μ202 = 167.39215085688076 - 155.83784914311926
if(μ201μ202>μ1μ2):
  print("μ1μ2較小=",μ1μ2)
else:
  print("μ201μ202較小=",μ201μ202)
```
